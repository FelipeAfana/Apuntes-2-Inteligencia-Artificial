```python
# prompt: crea un dataframe con el nombre, el cargo, el salario y la antiguedad en años, poner 20 filas de datos sinteticos

import pandas as pd
import numpy as np

# Crear datos sintéticos
np.random.seed(42)  # Para reproducibilidad
num_rows = 20
nombres = ['Nombre' + str(i) for i in range(1, num_rows + 1)]
cargos = np.random.choice(['Analista', 'Ingeniero', 'Gerente', 'Desarrollador'], size=num_rows)
salarios = np.random.randint(30000, 100000, size=num_rows)
antiguedades = np.random.randint(1, 10, size=num_rows)

# Crear el DataFrame
df = pd.DataFrame({
    'Nombre': nombres,
    'Cargo': cargos,
    'Salario': salarios,
    'Antiguedad': antiguedades
})

df
```

```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/adiacla/bigdata/refs/heads/master/venta_vivienda.csv",sep=";")
df
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(axis=0,inplace=True)
df
```

```python
df[["Latitud","Longitud","Area_construida_pie"]]
```

```python
df.info()
```

```python
# prompt: los features "Latitud","Longitud","Area_construida_pie" del df cambiar la coma por puntos

# Reemplazar comas por puntos en las columnas especificadas
for col in ["Latitud", "Longitud", "Area_construida_pie"]:
    df[col] = df[col].astype(str).str.replace(',', '.', regex=False)

# Convertir las columnas a tipo numérico (si es posible)
for col in ["Latitud", "Longitud", "Area_construida_pie"]:
    try:
      df[col] = pd.to_numeric(df[col])
    except ValueError:
      print(f"No se pudo convertir la columna '{col}' a numérico. Posibles valores no numéricos presentes.")

df.info()
```
